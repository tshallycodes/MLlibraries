```python
import numpy as np

# 1. Creating Arrays
arr = np.array([1, 2, 3])               # 1D array
matrix = np.array([[1, 2], [3, 4]])     # 2D array

# 2. Generate Arrays
zeros = np.zeros((2, 2))                # Array of zeros
ones = np.ones((2, 2))                  # Array of ones
full = np.full((2, 2), 7)               # Filled with a constant
arange = np.arange(0, 10, 2)            # Range of numbers
linspace = np.linspace(0, 1, 5)         # Evenly spaced numbers

# 3. Random Numbers
rand = np.random.rand(3)                # Random numbers between 0 and 1
randn = np.random.randn(3)              # Random from standard normal distribution
randint = np.random.randint(1, 10, 5)   # Random integers

# 4. Reshaping Arrays
reshape = np.arange(6).reshape(2, 3)    # Reshape to 2x3
flatten = reshape.flatten()             # Flatten to 1D

# 5. Basic Operations
add = np.add([1, 2], [3, 4])            # Add arrays
subtract = np.subtract([5, 7], [2, 4])  # Subtract arrays
multiply = np.multiply([2, 3], [4, 5])  # Multiply arrays
divide = np.divide([10, 20], [2, 4])    # Divide arrays

# 6. Statistical Functions
mean = np.mean([1, 2, 3, 4, 5])         # Mean
median = np.median([1, 2, 3, 4, 5])     # Median
std = np.std([1, 2, 3, 4, 5])           # Standard deviation
min_val = np.min([3, 1, 4])             # Minimum
max_val = np.max([3, 1, 4])             # Maximum

# 7. Indexing and Slicing
arr2 = np.array([10, 20, 30, 40, 50])
element = arr2[2]                      # Access element
slice_arr = arr2[1:4]                   # Slice array

# 8. Boolean Indexing
bool_idx = arr2 > 25                    # Boolean array
filtered = arr2[bool_idx]               # Filtered array

# 9. Matrix Operations
dot_product = np.dot([1, 2], [3, 4])    # Dot product
transpose = np.transpose([[1, 2], [3, 4]])  # Transpose

# 10. Aggregation Functions
sum_val = np.sum([1, 2, 3, 4, 5])       # Sum of elements
cumsum = np.cumsum([1, 2, 3])           # Cumulative sum
prod = np.prod([1, 2, 3, 4])            # Product of elements

# 11. Advanced Functions
unique_vals = np.unique([1, 2, 2, 3, 4]) # Unique values
sort_arr = np.sort([3, 1, 2])            # Sorting

# 12. Broadcasting
broadcast_sum = np.array([1, 2, 3]) + 5  # Add scalar to array

# 13. Clipping
clipped = np.clip([1, 5, 8, 10], 3, 7)   # Clip values between 3 and 7

# 14. Logical Operations
logical_and = np.logical_and([True, False], [False, True])

# 15. Mathematical Functions
sqrt_arr = np.sqrt([4, 9, 16])           # Square root
exp_arr = np.exp([1, 2, 3])              # Exponential
log_arr = np.log([1, np.e, np.e**2])     # Natural log
```

```python
import matplotlib.pyplot as plt

youtube_views1 = [234, 321, 677, 234, 212, 128, 50]
youtube_views2 = [212, 353, 566, 534, 969, 447, 343]
youtube_views3 = [267, 345, 689, 434, 4, 564, 232]

days_of_week = ("Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday")

plt.plot(days_of_week, youtube_views1, marker = 'o', label = "Week 1")
plt.plot(days_of_week, youtube_views2, marker = 'o', label = "Week 2")
plt.plot(days_of_week, youtube_views3, marker = 'o', label = "Week 3")
plt.title("Youtube Views Per Day")
plt.ylabel("Number of Views")
plt.xlabel("Days of the Week")
plt.grid("True")
plt.xticks(rotation = 90)
plt.legend(loc = "upper right")
plt.show()
```
